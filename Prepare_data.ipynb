```python
# cannot make this part a functino because of the different column names 
dictionary_commodities = {}
for contract in contract_names:
    # key: contract, value: dataframe with the data on that contract
    dictionary_commodities[contract] = pd.DataFrame(df[df['Market_and_Exchange_Names'] == contract])
    # put into the date in the appropriate form to sort  
    dictionary_commodities[contract]["Report_Date_as_MM_DD_YYYY"] = pd.to_datetime(\
                                                            dictionary_commodities[contract]["Report_Date_as_MM_DD_YYYY"])
    dictionary_commodities[contract].sort_values(by="Report_Date_as_MM_DD_YYYY", inplace=True)
    # make date column the index
    dictionary_commodities[contract].set_index("Report_Date_as_MM_DD_YYYY", inplace=True)
    dictionary_commodities[contract].index.name = ""
    # create a net_position column for money managers
    dictionary_commodities[contract]["Net_Positioning_MM"] = dictionary_commodities[contract]["M_Money_Positions_Long_ALL"]-\
    dictionary_commodities[contract]["M_Money_Positions_Short_ALL"]
    # check for missing values the selected column, since there are some 
    if (dictionary_commodities[contract]['Change_in_Open_Interest_All'].isnull().sum() != 0 or
    dictionary_commodities[contract]['Change_in_M_Money_Long_All'].isnull().sum() != 0 or
    dictionary_commodities[contract]['Change_in_M_Money_Short_All'].isnull().sum() != 0 or
    dictionary_commodities[contract]['M_Money_Positions_Spread_ALL'].isnull().sum() != 0): 
        print(contract) # print contract to know where the first values in the following columns are going
        # to be missing 
        dictionary_commodities[contract]['Change_in_Open_Interest_All'] = dictionary_commodities[contract]\
        ["Open_Interest_All"] - dictionary_commodities[contract]["Open_Interest_All"].shift(1)
        
        dictionary_commodities[contract]['Change_in_M_Money_Long_All'] = dictionary_commodities[contract]\
        ["M_Money_Positions_Long_ALL"] - dictionary_commodities[contract]["M_Money_Positions_Long_ALL"].shift(1)
        
        dictionary_commodities[contract]['Change_in_M_Money_Short_All'] = dictionary_commodities[contract]\
        ["M_Money_Positions_Short_ALL"] - dictionary_commodities[contract]["M_Money_Positions_Short_ALL"].shift(1)
        
        dictionary_commodities[contract]['Change_in_M_Money_Spread_All'] = dictionary_commodities[contract]\
        ["M_Money_Positions_Spread_ALL"] - dictionary_commodities[contract]["M_Money_Positions_Spread_ALL"].shift(1)

    # create a change of net_position column for money managers, do this here -- so, only contracts with missing values
    # get into the loop above
    dictionary_commodities[contract]["Change_in_Net_Positioning_MM"] = dictionary_commodities[contract]["Net_Positioning_MM"]-\
    dictionary_commodities[contract]["Net_Positioning_MM"].shift(1)
    # forward/backward fill where choosen in this order arbitrarily 
    # all the other missing values are going to be the Trader_ columns -- so use forward fill
    dictionary_commodities[contract].fillna(method='ffill', inplace=True)
    # fill in the remaining values
    dictionary_commodities[contract].fillna(method='bfill', inplace=True)
```

```python
def set_cftc_index(df, fill_method):
    use_index = load_obj("cftc_index")
    df = df.reindex(pd.date_range(use_index[0],use_index[-1], freq='D'))
    #df = df.asfreq("D")
    if fill_method == "ffill":
        df.fillna(method="ffill", inplace=True)
        # fill the remaining
        df.fillna(method="bfill", inplace=True)
    elif fill_method == "bfill":
        df.fillna(method="bfill", inplace=True)
        # fill the remaining
        df.fillna(method="ffill", inplace=True)
    elif fill_method == "avg":
        df = df.where(df==np.nan, other=(df.fillna(method="ffill")+df.fillna(method="bfill"))/2)
        # fill the remaining
        df.fillna(method="ffill", inplace=True)
        df.fillna(method="bfill", inplace=True)
    
    df = df.reindex(use_index)    
    return pd.DataFrame(df)
```

```python
def get_change_bool(df):
    # create difference column
    df["change"] = df - df.shift(1)
    # drop first row because nan is now present in the "change" column 
    df.drop(df.index[0], inplace=True)
    df["change_bool"] = 0
    # mark days with positive or no change as 0
    # mark days with negative change as 1
    df.loc[df["change"] < 0, "change_bool"] = 1
    # crop the data to the same time period as cftc data
    df = df.loc[start:end]
    return df

def get_index_neg_ret(df, n):
    df = get_change_bool(df)
    # parameter "n" is the number of days with negative return in a week
    # the function returns list of the dates when change was negative n days in a row 
    # resample by week
    df_resampled = df.resample("W").sum()["change_bool"]
    # list of dates of the beginings of the weeks
    list_begin = df_resampled.index[df_resampled == n] - pd.to_timedelta('6 days')
    # list of dates of the ends of the weeks 
    list_end = list_begin + pd.to_timedelta('4 days')
    ranges_list = []
    # create daterange and put in into list 
    for i in range(len(list_begin)):
        ranges_list.append(pd.date_range(start=list_begin[i], end=list_end[i], freq="D"))
    return ranges_list
```

```python
def get_indecies_cftc(neg_dates_list):
    # cftc data is reported on each Tuesday, so will consider Tuesdays, which are on the next week after
    # a week with certain number of days with negative returns.  
    indecies = []
    for i in range(len(neg_dates_list)):
        date = neg_dates_list[i][-1] + pd.to_timedelta("4 days")
        indecies.append(date)
    return indecies

```
